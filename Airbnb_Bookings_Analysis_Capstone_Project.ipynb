Since the  Hotel Name and Host Name have some Null values so we will remove all the rows with null because without name and hostel name we can cannot categorize the hostel.

```python
df.dropna(axis=0, subset=['name'], inplace=True)
df.dropna(axis=0, subset=['host_name'], inplace=True)
```

```python
df.isnull().sum()
```

```python
# plot before replacing the NULL value
sns.set(rc={'figure.figsize':(10.3,8.11)}) 
ax = sns.distplot(df['reviews_per_month'])
```

**Now, We have removed all the rows with Null values in Hostel Name and Host Name.We can also see the  Reviews_per_month has also have plenty of rows with Null values. We will relace all the NULL values with mean of that column.**

```python
mean_monthly_review= df['reviews_per_month'].mean()
df['reviews_per_month'].fillna(value=mean_monthly_review,inplace=True)
```

```python
#plotting after replacing the NULL value
sns.set(rc={'figure.figsize':(10.3,8.11)}) 
ax = sns.distplot(df['reviews_per_month'])
```

The only column remaining with Null values is last_month. we can replace the value with 0.  

```python
df.fillna(0, inplace=True)
df.isnull().sum()
```

### So we are done with dealing with Null values.

###Now we can see that columns like id, latitude, host_id and longitude doesn't make any sense at this point.so we will drop the these columns. 



```python
df.drop(['id','latitude','longitude','host_id'], axis=1, inplace=True)
```

```python
df.columns
```

### Now here we are seeing that the minimum price of the Hotel is Rs.0. which doesn't make any sense.. so we will replace the value with the average value of all hotel price. and also the max of Minimum night is too much.


```python
df['price']=df['price'].apply(lambda x: df['price'].mean() if x==0 else x)
```

```python
df['minimum_nights']=df['minimum_nights'].apply(lambda x: 365 if x>365 else x)
```

## how many properties lisitng is under a host..??


```python
host_listings= df.groupby('neighbourhood_group', as_index=True).agg({'host_name':'count','calculated_host_listings_count':'sum'}).reset_index()
host_listings.rename(columns={'host_name':'host_per_neighbour','calculated_host_listings_count':'no_of_lisitings_per_neighbourhood'},inplace=True)
host_listings
```

```python
sns.barplot(data=host_listings,y='host_per_neighbour',x='no_of_lisitings_per_neighbourhood',hue='neighbourhood_group')
```

## As we can see that when we group by the data on the basis of no of host per neighbour and  how many listings happend in that neighbour we  can see that there are 21643 hosts and 277044 no of listings followed by brooklyn with 20089 hosts and 45908. 

## And the Staten Island with 373	host and 865 host listing stands at the last in our dataset.


## let's try know more about the host listing dataframe.

```python
 colors = ['#ff9599','#66b7ff','#98ff99','#ffbc99']
fig1, ax1 = plt.subplots()
ax1.pie(host_listings['host_per_neighbour'],  labels=host_listings['neighbourhood_group'],  colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=90)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')
plt.tight_layout()
plt.show()
```
